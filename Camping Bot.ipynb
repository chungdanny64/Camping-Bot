```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import datetime
import time

driver = webdriver.Chrome()

driver.get('https://www.reservecalifornia.com/CaliforniaWebHome/Facilities/SearchViewUnitAvailabity.aspx')

dates = driver.find_element_by_class_name('ui-datepicker-trigger')

dates.click()

day = driver.find_element_by_link_text(#insert arrival Date)

day.click()

nights = driver.find_element_by_name('ctl01$mainContent$SearchUnitAvailbity$ddlNightsSearchUnitAvailbity')

nights.click()
    
#enter number of nights after option
numberofnights = driver.find_element_by_xpath('//*[@id="ddlNightsSearchUnitAvailbity"]/option[]') 

numberofnights.click()

place = driver.find_element_by_name('ctl01$mainContent$SearchUnitAvailbity$txtCityParkSearch')

place.clear()

place.send_keys('Millerton Lake SRA')
time.sleep(1)

placeclicker = driver.find_element_by_xpath('/html/body/ul/li/a')
placeclicker.click()

search = driver.find_element_by_class_name('expand_box')

search.click()




driver.find_element_by_xpath('//*[@id="aLogin"]/span').click()
wait = WebDriverWait(driver, 1)
element = wait.until(EC.element_to_be_clickable((By.ID, 'txtUserName')))
email = driver.find_element_by_id('txtUserName')
# insert username
email.send_keys(#)

password = driver.find_element_by_id('txtPassword')
# insert password
password.send_keys(#)

loginbutton = driver.find_element_by_xpath('//*[@id="divOnlyLogin"]')
loginbutton.click()

```

```python
sites = driver.find_element_by_xpath('/html/body/form/div[8]/main/div/div/div[10]/div/div[4]/div[3]/div/div[3]/table/tbody/tr[2]/td[2]/div/div[2]/a')
sites.click()


#enter the minute you want to wait until (will resume the bot exactly at a given time)

now = datetime.datetime.now()
while(now.minute != #):
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="PickTimeGridButton"]').click()
    now = datetime.datetime.now()
    
#replace with spot you want (goes in order)
wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="td_0_2"]')))
spot = driver.find_element_by_xpath('//*[@id="td_0_2"]')
spot.click()

wait.until(EC.presence_of_element_located((By.ID, 'mainContent_UnitDetailPopup_bReserveCalMobileCaptcha')))
time.sleep(.5)
driver.find_element_by_link_text('Reserve Unit').click()
wait.until(EC.presence_of_element_located((By.XPATH,'//*[@id="mainContent_ddlAdults"]')))
people = driver.find_element_by_xpath('//*[@id="mainContent_ddlAdults"]')
people.click()
nums = driver.find_element_by_xpath('//*[@id="mainContent_ddlAdults"]/option[4]')
nums.click()
unit_type = driver.find_element_by_xpath('//*[@id="mainContent_ddlSleepingUnitType"]')
unit_type.click()
num = driver.find_element_by_xpath('//*[@id="mainContent_ddlSleepingUnitType"]/option[5]')
num.click()

car = driver.find_element_by_xpath('//*[@id="mainContent_ddlPadLength"]')
car.click()
length = driver.find_element_by_xpath('//*[@id="mainContent_ddlPadLength"]/option[3]')
length.click()

name =  driver.find_element_by_xpath('//*[@id="mainContent_txtOccupant"]')
#enter Name
name.send_keys(#)

vehicle = driver.find_element_by_xpath('//*[@id="mainContent_ddlVehicles"]')
vehicle.click()
    
#enter number of vehicles after opotion
vehicle2 = driver.find_element_by_xpath('//*[@id="mainContent_ddlVehicles"]/option[]')
vehicle2.click()

agree = driver.find_element_by_xpath('/html/body/form/div[8]/main/div/div/div[2]/div[2]/div[2]')
agree.click()


plate = driver.find_element_by_xpath('//*[@id="ddlVehiclePlatNumber_1"]')
plate.click()

select= driver.find_element_by_xpath('//*[@id="ddlVehiclePlatNumber_1"]/option[2]')
select.click()
platenumber = driver.find_element_by_name('textbox1')
platenumber.send_keys('plate')

typer = driver.find_element_by_xpath('//*[@id="ddlVehicle_1"]')
typer.click()
clicker = driver.find_element_by_xpath('//*[@id="ddlVehicle_1"]/option[3]')
clicker.click()

reserve_unit = driver.find_element_by_xpath('//*[@id="mainContent_bReserve"]')
reserve_unit.click()

checkout = driver.find_element_by_xpath('//*[@id="mainContent_aCheckOut"]')
checkout.click()
```
